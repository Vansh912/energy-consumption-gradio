```python
final_df = data_encode.drop('Year', axis = 'columns')
```

```python
final_df = final_df.rename(columns={"Per Capita Energy Use (kWh)":'Per_Capita_Energy_Use',"Renewable Energy Share (%)":"Renewable_Energy_Share", "Fossil Fuel Dependency (%)":"Fossil_Fuel_Dependency", 
    "Industrial Energy Use (%)":"Industrial_Energy_Use", "Household Energy Use (%)":"Household_Energy_Use", 
    "Carbon Emissions (Million Tons)":"Carbon_Emissions","Energy Price Index (USD/kWh)":"Energy_Price_Index",'Total Energy Consumption (TWh)':"Total_Energy_Consumption"})
```

```python
final_df.head()
```

```python
Y = final_df['Total_Energy_Consumption']
```

```python
Y = (Y- Y.min()) / (Y.max() - Y.min())
```

```python
Y
```

```python
X = final_df.drop('Total_Energy_Consumption', axis = 'columns')
```

```python
X.head()
```

```python
for column in X.columns: 
    X[column] = (X[column] - X[column].min()) / (X[column].max() - X[column].min())
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
import xgboost as xgb
import numpy as np
from sklearn.metrics import mean_squared_error

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

params = {
    'objective': 'reg:squarederror',
    'eval_metric': 'rmse',
    'eta': 0.0009,
    'grow_policy': 'lossguide',
    'max_leaves': 256,
    'subsample': 0.5,
    'colsample_bytree': 0.8,
    'lambda': 1.0,
    'alpha': 0.9
}

evals = [(dtrain, 'train')]

bst = xgb.train(params, dtrain, num_boost_round=3000, evals=evals)

predictions = bst.predict(dtest)

rmse = np.sqrt(mean_squared_error(y_test, predictions))
print(f"Test RMSE: {rmse:.4f}")

```
